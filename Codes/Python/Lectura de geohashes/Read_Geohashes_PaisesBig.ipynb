# Paises grandes

```python
nompaises = ['Colombia','Argentina','Mexico','Brasil']
len(nompaises)
```

```python
zf = zipfile.ZipFile('Colombia.zip')
for i in zf.namelist():
    print(i)
```

```python
zf = zipfile.ZipFile('Colombia.zip')
df0 = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[0]))
df1 = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[1]))
df2 = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[2]))
df3 = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[3]))
df4 = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[4]))
df5 = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[5]))

df = pd.concat([df0, df1, df2, df3, df4, df5], ignore_index = True)
_, l = dask.compute(df, df.size)
print("Numeros de registros", l/ len(df.columns))
df.head(10)
```

```python
del zf, df0, df1, df2, df3, df4, df5, df, _, l
```
